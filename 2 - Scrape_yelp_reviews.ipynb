```python
url = "https://www.yelp.com/biz/arbys-montgomery-5?utm_campaign=yelp_api_v3&utm_medium=api_v3_business_search&utm_source=Hq7uCbsoGdMxeepn9RUHiA"
```

```python
from contextlib import nullcontext
import requests
from bs4 import BeautifulSoup
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
response = requests.get(url, headers=headers)
print(response.status_code)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all elements with class 'link'
    ratings_basic_info = soup.find_all(class_='css-1fdy0l5')
    reviews = soup.find_all(class_='comment__09f24__D0cxf css-qgunke')
    popular_dishes = soup.find_all(class_='css-wtz7x')


    # Extract and print the content of each 'link' element
    for basic_info in ratings_basic_info:
        print(basic_info.text)
        information_data['basic_info'] = basic_info.text
    for review in reviews:
        print(review.text)
        recommendation_data['reviews'] = review.text
    for dish in popular_dishes:
        print(dish.text)
        recommendation_data['popular_dishes'] = dish.text
else:
    information_data['basic_info'] = ''
    recommendation_data['reviews'] = ''
    recommendation_data['popular_dishes'] = ''

#rating and basic info:  css-1fdy0l5
#reviews: comment__09f24__D0cxf css-qgunke
#Popular dishes: css-wtz7x

```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import requests
from bs4 import BeautifulSoup
import time

def scrape_yelp_page(url):
    time.sleep(1)  # Sleep for 1 second before making a request
    r = {}

    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        r['reviews'] = [review.text for review in soup.find_all(class_='comment__09f24__D0cxf css-qgunke')]
        r['popular_dishes'] = [dish.text for dish in soup.find_all(class_='css-wtz7x')]
    else:
        r['reviews'] = ['']
        r['popular_dishes'] = ['']

    return url, r

# Function to process the results
def process_scraped_data(url, scraped_data):
    index = recommendation_data[recommendation_data['url'] == url].index[0]
    recommendation_data.at[index, 'reviews'] = scraped_data['reviews']
    recommendation_data.at[index, 'popular_dishes'] = scraped_data['popular_dishes']

# List of URLs to scrape
urls = recommendation_data['url'].tolist()
total_urls = len(urls)

# Number of threads to use (adjust as needed)
num_threads = 10

with ThreadPoolExecutor(max_workers=num_threads) as executor:
    # Submit scraping tasks to the executor
    future_to_url = {executor.submit(scrape_yelp_page, url): url for url in urls}

    # Initialize a counter for completed tasks
    completed_tasks = 0

    # Process results as they complete
    for future in as_completed(future_to_url):
        url = future_to_url[future]
        try:
            data = future.result()
            process_scraped_data(url, data[1])
            completed_tasks += 1
            print(f"Completed: {completed_tasks} out of {total_urls}")
        except Exception as exc:
            completed_tasks += 1
            print(f"Error for URL {url}: {exc}. Completed: {completed_tasks} out of {total_urls}")

print("All tasks completed.")
```
