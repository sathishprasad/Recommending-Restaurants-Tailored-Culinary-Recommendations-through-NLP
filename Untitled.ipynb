```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = 'https://www.yelp.com/biz/shanghai-durham?osq=Chinese'

health_score = []
# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the div with the specific class
    divs = soup.find_all('div', class_='arrange-unit__09f24__rqHTg arrange-unit-fill__09f24__CUubG css-1qn0b6x')
    amenities = [span.get_text() for div in divs for span in div.find_all('span')]
 # Print the text of each span
```

```python
amenities
```

```python
health_score = amenities
if 'Health Score' in health_score:
    index = health_score.index('Health Score')
    print(index)
```
